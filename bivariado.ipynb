```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
machines=pd.read_csv("./conjunto de datos/ALERTA.csv")
machines
```

```python
machines_num = machines.select_dtypes(include='number')
matriz = machines_num.corr()
print(matriz)

plt.matshow(matriz, cmap='bwr', vmin=-1, vmax=1)
plt.xticks(range(len(matriz.columns)), machines_num.columns, rotation=90)
plt.yticks(range(len(matriz.columns)), machines_num.columns)

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i,j, round(matriz.iloc[i,j], 2),ha='center', va='center')

plt.colorbar()
plt.show()
```

```python
#Relacion entre altura y peso ...
plt.scatter(machines["Air temperature [K]"],machines["Rotational speed [rpm]"])
plt.xlabel("Air temperature [K]")
plt.ylabel("Rotational speed [rpm]")
```

```python
#Correlacion Pearson

coeficiente=machines["Air temperature [K]"].corr(machines['Rotational speed [rpm]'])
print(f"coeficiente Pearson: {coeficiente}")
```
