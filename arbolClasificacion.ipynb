```python
#datos de entrenamiento - prueba
from sklearn.model_selection import train_test_split

datos_train, datos_test, clase_train, clase_test= train_test_split(
    pacientes[['edad','colesterol']],
    pacientes['problema_cardiaco'], test_size=0.30)
```

```python
from sklearn import tree #importando el modelo

arbol=tree.DecisionTreeClassifier(criterion='gini',max_depth=4) #Creando la instancia del modelo

modelo= arbol.fit(datos_train, clase_train)
print(tree.export_text(modelo, feature_names=['Edad', 'Colesterol']))

```

```python
plt.figure(figsize=(100,100))
tree.plot_tree(modelo,feature_names=['Edad','Colesterol'],class_names=['Sano','Cardiaco'],filled=True)
plt.show()
```

```python
exactitud= arbol.score(datos_test,clase_test)
print(exactitud)
```
